## Project: GA4 Merchandise Store Star Schema
Objective: Perform structural validation and quality assurance (QA) on the GA4 dataset before feature engineering and modeling. This notebook ensures referential integrity, schema adherence, and temporal consistency.

### Data Scope:

dim_users: User demographics and lifetime value.

dim_sessions: Session metadata, devices, and geography.

fct_events: Atomic event-level user interactions.

fct_product_interactions: Unnested product-level SKU data.

Status: Phase 1 — Structural Validation

```python
import pandas as pd
import os

# --- 1. CONFIGURATION & LOADING ---
def run_data_quality_audit(data_path='./'):
    # Define schema explicitly to avoid ID corruption (string format)
    dtypes = {'user_id': str, 'session_id': str, 'event_id': str, 'transaction_id': str}
    
    # Load files
    try:
        df_users = pd.read_csv(os.path.join(data_path, 'dim_users.csv'), dtype=dtypes)
        df_sessions = pd.read_csv(os.path.join(data_path, 'dim_sessions.csv'), dtype=dtypes)
        df_events = pd.read_csv(os.path.join(data_path, 'fct_events.csv'), dtype=dtypes)
        df_products = pd.read_csv(os.path.join(data_path, 'fct_product_interactions.csv'), dtype=dtypes)
    except FileNotFoundError as e:
        print(f"Error: Could not find file. Check your data_path. Details: {e}")
        return None

    print("--- DATA QUALITY REPORT ---")

    # --- 2. REFERENTIAL INTEGRITY (Star Schema Health) ---
    # Every event must have a parent session.
    orphans = df_events[~df_events['session_id'].isin(df_sessions['session_id'])]
    print(f"1. Referential Integrity: {len(orphans)} orphaned events found (Target: 0)")

    # --- 3. UNIQUENESS (Primary Key Check) ---
    # event_id must be unique. If not, your counts will be inflated.
    dupes = df_events['event_id'].duplicated().sum()
    print(f"2. Uniqueness Check: {dupes} duplicate event IDs found (Target: 0)")

    # --- 4. COMPLETENESS (Null/Missing Values) ---
    # Check for empty user_ids, which make cohorts impossible to track.
    null_users = df_events['user_id'].isnull().sum()
    print(f"3. Completeness Check: {null_users} events missing user_id (Target: 0)")

   # --- 5. CONSISTENCY (Temporal Logic) ---
    # We use format='mixed' to handle inconsistent precision (some rows have microseconds, some don't)
    df_sessions['start'] = pd.to_datetime(df_sessions['session_start_time'], format='mixed')
    df_sessions['end'] = pd.to_datetime(df_sessions['session_end_time'], format='mixed')
    
    # Now check for temporal drift
    invalid_sessions = df_sessions[df_sessions['end'] < df_sessions['start']]
    print(f"4. Temporal Consistency: {len(invalid_sessions)} corrupted session times found (Target: 0)")

    # --- 6. VALIDITY (Sanity Check) ---
    # Ensure revenue doesn't result in NaN or impossible values.
    total_rev = df_events['revenue_usd'].fillna(0).sum()
    print(f"5. Financial Validity: Total Revenue identified = ${total_rev:,.2f}")

    return df_users, df_sessions, df_events, df_products

# --- EXECUTION ---
# Change './' to your specific directory if needed
users, sessions, events, products = run_data_quality_audit(data_path='./')
```

### Loading these tables into Postgresql in local file

```python
import pandas as pd
from sqlalchemy import create_engine, types, text

# 1. DATABASE CONFIGURATION
# Replace the connection string with your actual Postgres credentials
DB_URL = "postgresql://shekhar:admin123@localhost:5432/saas_analytics"
engine = create_engine(DB_URL)
SCHEMA_NAME = "ga4_analytics"

# 2. CONFIGURATION MAP
# This map explicitly defines the types for every column you provided.
PIPELINE_CONFIG = {
    'dim_users.csv': {
        'table_name': 'dim_users',
        'dtype': {
            'user_id': types.VARCHAR(255),
            'first_seen_date': types.Date(),
            'last_seen_date': types.Date(),
            'total_sessions': types.Integer(),
            'lifetime_value_usd': types.Numeric(10, 2)
        },
        'date_cols': ['first_seen_date', 'last_seen_date']
    },
    'dim_sessions.csv': {
        'table_name': 'dim_sessions',
        'dtype': {
            'session_id': types.VARCHAR(255),
            'user_id': types.VARCHAR(255),
            'session_start_time': types.DateTime(),
            'session_end_time': types.DateTime(),
            'device_type': types.VARCHAR(100),
            'operating_system': types.VARCHAR(100),
            'country': types.VARCHAR(100),
            'city': types.VARCHAR(100),
            'start': types.DateTime(),
            'end': types.DateTime()
        },
        'date_cols': ['session_start_time', 'session_end_time', 'start', 'end']
    },
    'fct_events.csv': {
        'table_name': 'fct_events',
        'dtype': {
            'event_id': types.VARCHAR(255),
            'session_id': types.VARCHAR(255),
            'user_id': types.VARCHAR(255),
            'event_time': types.DateTime(),
            'event_name': types.VARCHAR(100),
            'page_url': types.Text(),
            'page_title': types.Text(),
            'transaction_id': types.VARCHAR(255),
            'revenue_usd': types.Numeric(10, 2)
        },
        'date_cols': ['event_time']
    },
    'fct_product_interactions.csv': {
        'table_name': 'fct_product_interactions',
        'dtype': {
            'interaction_id': types.VARCHAR(255),
            'session_id': types.VARCHAR(255),
            'event_time': types.DateTime(),
            'event_name': types.VARCHAR(100),
            'item_id': types.VARCHAR(100),
            'item_name': types.Text(),
            'item_category': types.VARCHAR(100),
            'item_price': types.Numeric(10, 2),
            'item_quantity': types.Integer()
        },
        'date_cols': ['event_time']
    }
}

# 3. ETL EXECUTION
def run_etl(config):
    # Establish connection
    with engine.connect() as conn:
        # Create schema using 'text' to comply with SQLAlchemy 2.0+
        with conn.begin():
            conn.execute(text(f"CREATE SCHEMA IF NOT EXISTS {SCHEMA_NAME};"))
    
    for filename, settings in config.items():
        try:
            print(f"Loading {filename} into {settings['table_name']}...")
            
            # Load CSV, strictly as string to prevent ID truncation
            df = pd.read_csv(filename, dtype=str)
            
            # Process Date Columns
            for col in settings['date_cols']:
                if col in df.columns:
                    df[col] = pd.to_datetime(df[col], format='mixed', errors='coerce')
            
            # Load into PostgreSQL
            df.to_sql(
                settings['table_name'],
                engine,
                schema=SCHEMA_NAME,
                if_exists='replace',
                index=False,
                dtype=settings['dtype']
            )
            print(f"Success: {settings['table_name']} is now in Postgres.")
            
        except Exception as e:
            print(f"Error loading {filename}: {e}")

if __name__ == "__main__":
    run_etl(PIPELINE_CONFIG)
    print("\nETL Pipeline Complete.")
```

```python
import pandas as pd
from sqlalchemy import create_engine, types, text

# 1. DATABASE CONFIGURATION
DB_URL = "postgresql://shekhar:admin123@localhost:5432/saas_analytics"
engine = create_engine(DB_URL)
SCHEMA_NAME = "ga4_analytics"

def run_etl_refined():
    filename = 'fct_product_interactions.csv'
    table_name = 'fct_product_interactions'
    
    # 1. Load data
    df = pd.read_csv(filename, on_bad_lines='warn', low_memory=False)
    
    # 2. Forensic Cleaning
    # Convert to numeric, turn errors (text) into NaN
    df['item_quantity'] = pd.to_numeric(df['item_quantity'], errors='coerce')
    
    # Fill missing values (NaN) with 1
    df['item_quantity'] = df['item_quantity'].fillna(1)
    
    # Clip outliers: Anything over 999 becomes 999
    # This prevents the 'NumericValueOutOfRange' error and keeps data sane
    df['item_quantity'] = df['item_quantity'].clip(upper=999)
    
    # Convert to int
    df['item_quantity'] = df['item_quantity'].astype(int)
    
    print(f"Data cleaned. Rows: {len(df)}")
    
    # 3. Load
    # We define the type as Integer (it will now fit, as max is 999)
    df.to_sql(table_name, engine, schema=SCHEMA_NAME, if_exists='replace', index=False)
    print("ETL complete. Data is now in Postgres.")

if __name__ == "__main__":
    run_etl_refined()
```
